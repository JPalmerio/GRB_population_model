```python
import sys
src_dir = '/Users/palmerio/Science_projects/GRB_population_model/grbpop/'
sys.path.insert(0, src_dir)
from GRB_population import GRBPopulation, create_GRB_population_from
import physics as ph
import numpy as np
import io_grb_pop as io
import miscellaneous as msc
from cosmology import init_cosmology, Lum_dist
from ECLAIRs import init_ECLAIRs
import plotting_functions as pf
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import logging
import f90_functions as f90f

log = logging.getLogger(__name__)
logging.basicConfig(stream=sys.stdout, level=logging.DEBUG,
                    format='%(asctime)s.%(msecs)03d [%(levelname)s] %(message)s',
                    datefmt='%H:%M:%S')
logging.getLogger('matplotlib').setLevel(logging.WARNING)
plt.style.use('presentation')
```

```python
paths_to_dir, paths_to_files = io.generate_paths()
```

```python
config, params, instruments, samples, obs_constraints = io.read_init_files(paths_to_files)
config['Nb_GRBs'] = '1e5'
config['samples'] = ['Stern']
config['constraints'] = ['Stern']
config['output_dir'] = 'debugger'
# paths_to_dir['output'] = paths_to_dir['output']/config['output_dir']
paths_to_dir['output']=Path('/Users/palmerio/Science_projects/GRB_population_model/model_outputs/debugger/')
```

```python
cosmo = init_cosmology(paths_to_dir['cosmo'])
ECLAIRs_prop = init_ECLAIRs(ECLAIRs_dir=paths_to_dir['ECLAIRs'],
                            ECLAIRs_config=instruments['ECLAIRs'])
samples['ECLAIRs']['pflx_min'] = ECLAIRs_prop['bkg_total']

incl_samples, incl_instruments, incl_constraints = io.create_config(config=config,
                                                       samples=samples,
                                                       instruments=instruments,
                                                       obs_constraints=obs_constraints)
```

```python
params['luminosity_function']['model'] = 'Fixed'
params['luminosity_function']['logL0'] = 52.
params['spectral_shape']['model'] = 'Fixed'
params['spectral_shape']['shape'] = 'Band'
np.random.seed(12)
gp = create_GRB_population_from(Nb_GRBs=config['Nb_GRBs'],
                                cosmo=cosmo,
                                params=params,
                                incl_samples=incl_samples,
                                incl_instruments=incl_instruments,
                                incl_constraints=incl_constraints,
                                ECLAIRs_prop=ECLAIRs_prop,
                                output_dir=paths_to_dir['output'])
```

```python
params['spectral_shape']['shape'] = 'BPL'
np.random.seed(12)
gp2 = create_GRB_population_from(Nb_GRBs=config['Nb_GRBs'],
                                cosmo=cosmo,
                                params=params,
                                incl_samples=incl_samples,
                                incl_instruments=incl_instruments,
                                incl_constraints=incl_constraints,
                                ECLAIRs_prop=ECLAIRs_prop,
                                output_dir=paths_to_dir['output'])
```

```python
plt.style.use('paper')
# fig, ax = plt.subplots(figsize=(10,8))

gp.properties.sort_values(by='z', inplace=True)
gp2.properties.sort_values(by='z', inplace=True)
fig, axes = pf.cool_hist2d(x=np.log10(gp.properties['pht_pflx_BATSE']),
                           y=np.log10(gp2.properties['pht_pflx_BATSE']),
                           c=gp.properties['z'], s=10, linewidth=0, cblabel='Redshift (z)',
                           mode='scatter')
axes['center'].plot(np.linspace(-6,4),np.linspace(-6,4),color='k', zorder=0)
axes['center'].plot(np.linspace(-6,4),np.linspace(-6,4)-1,color='k', zorder=0, ls='--', lw=1)
axes['center'].axvline(-1, ls=':',color='k', lw=2, zorder=0)
axes['center'].set_xlabel('log pht pflx Band')
axes['left'].set_ylabel('log pht pflx Broken Power Law')

fig, axes = plt.subplots(2,2, figsize=(10,10), tight_layout=True)
axes[0,0].hist(gp.properties['alpha'], bins=np.linspace(-1,2,31), color='grey', density=True, histtype='step', linewidth=3)
axes[0,0].hist(gp2.properties['alpha'], bins=np.linspace(-1,2,31), color='C3', density=True, histtype='step', linewidth=3)
axes[0,1].hist(gp.properties['beta'], bins=np.linspace(2,5,31), color='grey', density=True, histtype='step', linewidth=3)
axes[0,1].hist(gp2.properties['beta'], bins=np.linspace(2,5,31), color='C3', density=True, histtype='step', linewidth=3)
axes[1,0].hist(gp.properties['ktild'], bins=np.linspace(0,5,51), color='grey', density=True, histtype='step', linewidth=3,
            label='Band')
axes[1,0].hist(gp2.properties['ktild'], bins=np.linspace(0,5,51), color='C3', density=True, histtype='step', linewidth=3,
            label='BPL')
axes[1,1].hist(np.log10(gp.properties['pht_pflx_BATSE']), bins=np.linspace(-5,4,91), color='grey', density=True, histtype='step', linewidth=3)
axes[1,1].hist(np.log10(gp2.properties['pht_pflx_BATSE']), bins=np.linspace(-5,4,91), color='C3', density=True, histtype='step', linewidth=3)

axes[1,0].legend()
axes[0,0].set_xlabel('Alpha')
axes[0,1].set_xlabel('Beta')
axes[1,0].set_xlabel('ktild')
axes[1,1].set_xlabel('log pht pflx BATSE')
plt.show()
```

```python
plt.style.use('paper')
fig, ax = plt.subplots(figsize=(10,8))
pf.plot_intensity_constraint(ax=ax, pop=gp, label='Tsim')
ax.legend()
fig, ax = plt.subplots(figsize=(10,8))
pf.plot_redshift_constraint(ax=ax, pop=gp, label='Tsim')
ax.legend()
```
